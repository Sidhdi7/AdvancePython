## 4. Read CSV in chunks (memory-efficient)

```python
chunk_iter = pd.read_csv(r"C:\Users\ADMIN\Downloads\house_data (1).csv", chunksize=5000)
for i, chunk in enumerate(chunk_iter):
    print(f"Chunk {i+1} -> shape: {chunk.shape}")
    display(chunk.head(2))
    # For demo, break after printing the first chunk
    if i >= 0:
        break
```

## 5. Create numeric-only and text-only CSVs

```python
# Numeric-only file
numeric_cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'sqft_living15', 'sqft_lot15']
print('Numeric columns:', numeric_cols)

num_df = house_df[numeric_cols]
num_df.head()

# Text-only file
text_cols = ['date', 'zipcode']
print('\nText columns:', text_cols)

text_df = house_df[text_cols]
text_df.head()
```

## 6. Quick examples: filtering and simple plot

```python
# Filter houses with price > 500k and bedrooms >=3
filtered = house_df[(house_df['price'] > 500000) & (house_df['bedrooms'] >= 3)]
filtered[['id','price','bedrooms','sqft_living']].head()
```
